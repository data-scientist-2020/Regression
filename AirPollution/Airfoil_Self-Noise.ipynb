This is a regression Project which helps in determine the Scaled sound pressure level in air

```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import probplot
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
#Load the data
data = pd.read_csv('airfoil_self_noise.dat', sep='\t', names=['Frequency','Angle of attack','Chord length',
                                                'Free-stream velocity','Suction side displacement thickness',
                                                              'Scaled sound pressure level'])
```

```python
#Visualise the data using pair plot
sns.pairplot(data)
```

```python
#Visulaise the data using heat map
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True)
```

```python
#Seggregate the column to X and y
X = data[['Frequency','Angle of attack','Chord length','Free-stream velocity',
          'Suction side displacement thickness']]
y = data['Scaled sound pressure level']
```

```python
#Split the data to training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#Method to visualise the distribution of the data
def diagonastic_plot(data,column):
    
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    sns.distplot(data[column])
    
    plt.subplot(1,2,2)
    probplot(data[column],plot=plt)
    
    plt.show()
```

```python
#Visualise the data distribution
for col in X_train.columns:
    diagonastic_plot(X_train,col)
```

```python
#Model selection based on kernel
kernel_dict = {}
for kernel in ['linear','rbf','poly']:
    pipeline_k = Pipeline([
        ('scaler', StandardScaler()),
        ('model', SVR(kernel=kernel))
    ])
    score_k = cross_val_score(estimator=pipeline_k,X=X_train,y=y_train,cv=10,scoring='r2')
    kernel_dict[kernel] = score_k.mean()
```

```python
kernel_dict
```

```python
#Model selection based on Gamma
gamma_dict = {}
for i in [0.90,0.91,0.92,0.93,0.94,1.0,2.0]:
    pipeline_g = Pipeline([
        ('scaler', StandardScaler()),
        ('model', SVR(kernel = 'rbf', gamma=i ))
    ])
    score_g = cross_val_score(estimator=pipeline_g, X=X_train, y=y_train,cv=10,scoring='r2')
    gamma_dict[i] = score_g.mean()
```

```python
gamma_dict
```

```python
c_dict = {}
for i in range(1,10):
    pipeline_c = Pipeline([
        ('scaler', StandardScaler()),
        ('model', SVR(kernel='rbf', gamma=0.9, C=i))
    ])
    score_c = cross_val_score(estimator=pipeline_c, X=X_train, y=y_train, cv=10, scoring='r2')
    c_dict[i] = score_c.mean()
```

```python
c_dict
```

```python
#Standardize the data
scaler = StandardScaler()
scaler.fit(X_train, y_train)
```

```python
#transform the data
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Model Buiding
regressor = SVR(kernel='rbf', gamma=0.9, C=15)
regressor.fit(X_train,y_train)
```

```python
#Evaluate the model using cross_val_score on test data
score = cross_val_score(estimator=regressor, X=X_test, y=y_test, cv=9, scoring='r2', n_jobs=-1)
```
